```python
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_excel('10 days analysis of my data.xlsx')
```

```python
data.head()
data.columns
data.shape
data.isnull().values.any()
data.isnull().sum()
```

```python
data.describe
```

```python

# Stripping extra spaces in the description 
data['Order Item Name'] = data['Order Item Name'].str.strip() 
  
# Dropping the rows without any invoice number 
data.dropna(axis = 0, subset =['Customer Phone'], inplace = True) 
data['Customer Phone'] = data['Customer Phone'].astype('str') 
  
data['Order Item Name'] = data['Order Item Name'].str.strip() #removes spaces from beginning and end
data.dropna(axis=0, subset=['Customer Phone'], inplace=True) #removes duplicate invoice
data['Customer Phone'] = data['Customer Phone'].astype('str') #converting invoice number to be string

data.head()
```

```python


basket_France = (data[data['Order Status'] =="completed"] 
          .groupby(['Customer Phone', 'Order Item Name'])['Order Item Qt Y']      
          .sum().unstack().reset_index().fillna(0) 
          .set_index('Customer Phone'))
basket_France


```

```python
def hot_encode(x): 
    if(x<= 0): 
        return 0
    if(x>= 1): 
        return 1

basket_encoded = basket_France.applymap(hot_encode) 
basket_France = basket_encoded

basket_France
```

```python

my_frequent_itemsets = apriori(basket_France, min_support=0.0007, use_colnames=True)
my_frequent_itemsets
```

```python
my_rules = association_rules(my_frequent_itemsets, metric="lift", min_threshold=1)
```

```python
my_rules.head(100)
datatoexcel=pd.ExcelWriter("Surid Apriori Algo results.xlsx",engine='xlsxwriter')

my_rules.head(100).to_excel(datatoexcel,sheet_name='Sheet1')
datatoexcel.save()
```
